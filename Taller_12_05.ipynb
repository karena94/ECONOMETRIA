```python
!pip install wooldridge
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import wooldridge as wd
import statsmodels.api as sm
import statsmodels.stats.api as sms
```

##Primero

```python
df=wd.data("sleep75") #cargar tabla
df
```

```python
wd.data("sleep75",description=True) #age,educ,gdhlth,inlf,leis1,union,exper,yngkid,yrsmarr
```

```python
X=df[["age","educ","gdhlth","inlf","leis1","union","exper","yngkid","yrsmarr"]]
X=sm.add_constant(X)
X.head()
```

```python
y=df["sleep"]
```

```python
model=sm.OLS(y,X).fit()
model.summary
model
```

```python
sms.het_white(model.resid,model.model.exog)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                
```

```python
sms.het_breuschpagan(model.resid,model.model.exog)
```

Con las dos pruebas se puede afirmar que el modelo es HETEROCEDASTICO, porque el Pvalor se encuentra por debajo. Se rechaza la hipotesis nula, asi que las variables afectan los minutos de sueño por semana.
